looking for missing values

```python
missing_value = df.isnull().sum()

print(missing_value)
```

### filling up the missing values 
Let's fill in missing values in the DataFrame with the median value of each NUMERIC column

```python
df.fillna(df.mean(numeric_only=True).round(1), inplace=True)
df.info()
```

 now lets fill in missing values in the selected string columns using the most frequent value of each respective column

```python
string_columns = df.select_dtypes(include=['object']).columns
df[string_columns] = df[string_columns].fillna(df[string_columns].mode().iloc[0]) 
```

```python
# lets view our datafram after filling missing values 
missing_value1 = df.isnull().sum()

print(missing_value1)
```

Making some necessary changes on the detaframe 

```python
# lets change the data type of the "mileage" in the dataFrame to integer and get rid of the decimals 

df = df.astype({'mileage':'int'})
```

### Replacing the binary (boolean) values in a more understandable methods.

```python
# converting specific values in the "accidents_or_damage" and "one_owner " column to more meaningful labels, 
# after replacing 1 with 'Yes' and 0 with 'No', this will make our datafram more understandable. 

df['accidents_or_damage'] = df['accidents_or_damage'].replace({1: 'TRUE', 0: 'FALSE'})


```

```python
# similarly "one_owner" columns replacing with "true " or "false" 

df['one_owner'] = df['one_owner'].replace({1: 'TRUE', 0: 'FALSE'})
```

replacing the value in personal_use_only column that indicates if the car is for personal use or not into "TRUE" or "FALSE"

```python

# replacing '1' with 'True' and '0' with 'False'
df['personal_use_only'] = df['personal_use_only'].replace({1: 'TRUE', 0: 'FALSE'})
```
